```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

data = {
    "Solar":[1200,1500,1300],
    "Wind":[1000,1200,1100],
    "Hydropower":[2900,3100,2000],
    "Biomass": [2500, 2700,2400]
}
df = pd.DataFrame(data)
df.head()
sns.pairplot(df)
plt.show()
```

```python
#create a heatmap to show the correlation matrix
correlation_matrix = df.corr()

sns.heatmap(correlation_matrix, annot =True, cmap = 'coolwarm', linewidth = 0.5)

#add title
plt.title('Correlation betweeen Energy Resources')
plt.show()
```

```python
#create boxplot 
sns.boxplot(data=df)

#add title
plt.title('Distribution of Energy Consumption by Source')
plt.show()
```

```python
#violin plot is used when one data is continuous and other one is categorical data
sns.violinplot(data=df)
plt.title('Distribution of energy consumption by source')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

energy_values = [100,200,300,400,500]
carbon_emissions = [10,25,34,40,50]

df_reg = pd.DataFrame({
    'Energy Consumptions': energy_values,
    'Carbon Emissions': carbon_emissions
})

sns.regplot(x='Energy Consumptions', y='Carbon Emissions', data=df_reg)
```

```python
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt

fig = plt.figure()

#syntax for 3d projection
ax = plt.axes(projection = '3d')

#defining all 3 axes
z = np.linspace(0, 1, 1000)
x = z * np.sin(25 * z)
y = z * np.cos(25 * z)

#plotting
ax.plot3D(x, y, z, 'green')
ax.set_title('3D line plot')
plt.show()
```
